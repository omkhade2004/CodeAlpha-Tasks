```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')
```

```python
df = pd.read_csv('quotes_dataset (1).csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
top_authors = df['author'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_authors.plot(kind='bar')
plt.title('Top 10 Authors by Number of Quotes')
plt.xlabel('Author')
plt.ylabel('Number of Quotes')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(10,5))
sns.countplot(
    y=df['quote'],
    order=df['quote'].value_counts().head(10).index
)
plt.title('Top 10 Quote')
plt.show()
```

```python
df['quote_length'] = df['quote'].astype(str).apply(len)

plt.figure(figsize=(8,5))
plt.hist(df['quote_length'], bins=30)
plt.title('Distribution of Quote Lengths')
plt.xlabel('Quote Length')
plt.ylabel('Frequency')
plt.show()
```
